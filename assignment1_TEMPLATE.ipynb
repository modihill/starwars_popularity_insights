```python
#import libraries 
#pandas for data retieval and other basic functions
#matplotlib.pyplot as plt for generating graphs and charts

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data Retriving

#import data_set StarWars.csv as starWars
#skiping 1st row from starWars.csv as it only have responce or header value 

with open('StarWars.csv') as starWars_csv:
    starWars = pd.read_csv(starWars_csv, sep=',', decimal='.',skiprows=[1])
    
```

```python
#many columns header has value like unnamed:
#or have very long name like "Which of the following Star Wars films have you seen? Please select all that apply."
#for ease of understanding, I am renaming those columns with appropriate names

starWars = starWars.rename(columns = {
    'Which of the following Star Wars films have you seen? Please select all that apply.':'Seen Episode I',
    'Unnamed: 4':'Seen Episode II',
    'Unnamed: 5':'Seen Episode III',
    'Unnamed: 6':'Seen Episode IV',
    'Unnamed: 7':'Seen Episode V',
    'Unnamed: 8':'Seen Episode VI',
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'Rank of Episode I',
    'Unnamed: 10':'Rank of Episode II',
    'Unnamed: 11':'Rank of Episode III',
    'Unnamed: 12':'Rank of Episode IV',
    'Unnamed: 13':'Rank of Episode V',
    'Unnamed: 14':'Rank of Episode VI',
    'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.':'Han Solo character favorably, unfavorably, or unfamiliar',
    'Unnamed: 16':'Luke Skywalker character favorably, unfavorably, or unfamiliar',
    'Unnamed: 17':'Princess Leia Organa character favorably, unfavorably, or unfamiliar',
    'Unnamed: 18':'Anakin Skywalker character favorably, unfavorably, or unfamiliar',
    'Unnamed: 19':'Obi Wan Kenobi character favorably, unfavorably, or unfamiliar',
    'Unnamed: 20':'Emperor Palpatine character favorably, unfavorably, or unfamiliar',
    'Unnamed: 21':'Darth Vader character character favorably, unfavorably, or unfamiliar',
    'Unnamed: 22':'Lando Calrissian character favorably, unfavorably, or unfamiliar',
    'Unnamed: 23':'Boba Fett character favorably, unfavorably, or unfamiliar',
    'Unnamed: 24':'C-3P0 character favorably, unfavorably, or unfamiliar',
    'Unnamed: 25':'R2-D2 character favorably, unfavorably, or unfamiliar',
    'Unnamed: 26':'Jar Jar Binks character favorably, unfavorably, or unfamiliar',
    'Unnamed: 27':'Padme Amidala character favorably, unfavorably, or unfamiliar',
    'Unnamed: 28':'Yoda character favorably, unfavorably, or unfamiliar',
})

starWars.head(None)
```

### Task 1.3 : Typos


```python
#I have observe every columns with value_counts(), found some typos and whitespace, that i am going to solve with below codes


#Handling Whitespace of multiple columns
#with strip() function

starWars = starWars.applymap(lambda s:s.strip() if type(s) == str else s)
```

### Task 1.4 : Extra - whitespaces


```python
#handling typos of multiple columns
#as many columns have Yess, yes instead of Yes, 
#or F instead of Female and so on, 
#with replace() function

starWars = starWars.apply(lambda x: x.astype(str)
                .replace({
                    'nan': np.nan,
                    "Yess": "Yes",
                    "yes": "Yes",
                    "no": "No",
                    "Noo": "No",
                    "F": "Female",
                    "female": "Female",
                    "male": "Male",
                    })
                
                if x.dtype == "object" 
                else x
               )
```

### Task 1.5: Upper/Lower-case


```python
#Uppercase Conversion to whole dataset
#with upper() function

starWars = starWars.applymap(lambda u:u.upper() if type(u) == str else u)
```

### Task 1.6: Sanity checks


```python
#sanity Check

#with value_counts()
#I have observe every columns, and to conclude that Age column needs sanity check
#it has impossible value 500 (Outlier)
#that i am checking with for loop, enumerate and if condition, and going to replace outlier with maximum count of age - '45 - 60'
#that i am transforming with replace function

starWars['Age'].value_counts()

for i, value in enumerate(starWars['Age']):
    if value == '500':
        starWars['Age'][i] = '45-60'
        
starWars['Age'].value_counts()
```

### Task 1.7: Missing values

```python
#missing values

#dataset starWars have lots of null values that, if we drop them then dataset will become small and will loose some important observations
#that outcome of dataset will reduce its trustworthiness
#replacing null with mean or median will may lead to false estimates


#although i am replacing null values of following columns with no, assuming NaN means
#respondants are not fan of star wars franchise and not familier with expanded universe
starWars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = starWars['Do you consider yourself to be a fan of the Star Wars film franchise?'].fillna('NO')
starWars['Are you familiar with the Expanded Universe?'] = starWars['Are you familiar with the Expanded Universe?'].fillna('NO')
```

```python
#missing values

#replacing all NaN value in Seen Episode columns with no, assuming Respondent have not seen them
#if Seen Episodes columns have value, then relacing value with yes 


seen_episode_I = starWars.iloc[0]['Seen Episode I']
seen_episode_II = starWars.iloc[0]['Seen Episode II']
seen_episode_III = starWars.iloc[0]['Seen Episode III']
seen_episode_IV = starWars.iloc[0]['Seen Episode IV']
seen_episode_V = starWars.iloc[0]['Seen Episode V']
seen_episode_VI = starWars.iloc[0]['Seen Episode VI']

starWars = starWars.apply(lambda s: s.astype(str).replace({
    seen_episode_I: "YES",
    seen_episode_II: "YES",
    seen_episode_III: "YES",
    seen_episode_IV: "YES",
    seen_episode_V: "YES",
    seen_episode_VI: "YES",
    'nan': "NO"
    }) 
                          
        if s.name in [
        'Seen Episode I',
        'Seen Episode II',
        'Seen Episode III',
        'Seen Episode IV',
        'Seen Episode V',
        'Seen Episode VI'
    ] 
        else s)

starWars
```

## Task 2 : Data Exploration


### Task 2.1: Explore a survey question


```python
#Explore the Survey Question,
#how people analysis the survey question
#i am creating list [median_of_ranking] that includes median values of ranking of every star wars episodes


ranking_edpisodes = [
    'Rank of Episode I', 
    'Rank of Episode II', 
    'Rank of Episode III', 
    'Rank of Episode IV', 
    'Rank of Episode V', 
    'Rank of Episode VI'
]

median_of_ranking = []

for i, value in enumerate(ranking_edpisodes):   
    median_of_ranking.append(starWars[value].median())

median_of_ranking
```

```python
#then creating data frame called median_Ranking_df with median list of all ranking columns
#and plotting bar graph of that data frame

#here, we can conclude that, Episode III has highest median values, means it is ranked worst by respondents
#Episode V has lowest value, means it is ranked best by repondents


median_Ranking_df = pd.DataFrame([median_of_ranking], columns = ranking_edpisodes)

rank_chart = median_Ranking_df.plot(kind='bar', figsize=(10,10), fontsize = 20, edgecolor='black', color=('purple','skyblue','red','pink','green','orange'), stacked=False)
plt.title('Median Ranking per Star Wars Episodes', fontsize=20)
rank_chart.set_xlabel('Median value of Ranking of Episodes', fontsize=20)
rank_chart.set_ylabel("Rank",fontsize=20)
rank_chart.set_ylim(0,6)
plt.grid(color='grey', linewidth=3, linestyle='dotted', axis='y')
plt.show()
```

### Task 2.2: Relationships between columns

```python
#exploring relationship between columns

#relation 1
#bar graph of seen any movie from star wars film grouped by gender

#here, we can conclude that, number of females and males are approximately same when we talked about at least seen one movie 
#of star Wars franchise
#while number of females are much higher than males when we talked about have not seen atleast one movie.

#both male and female have equally seen any of 6 movie from star Wars franchise (male > 400, female slighly less than < 400)

relation_chart1 = starWars.groupby(['Gender',
             'Have you seen any of the 6 films in the Star Wars franchise?']).size().unstack().plot(kind='bar', edgecolor='black', legend=True, color=('orange', 'skyblue'),figsize=(10,10), fontsize=12)
plt.title('Seen any film of Star Wars Franchise by Gender', fontsize=20)
relation_chart1.set_xlabel('Gender', fontsize=20)
relation_chart1.set_ylabel("Frequency",fontsize=20)
plt.grid(color='grey', linewidth=3, linestyle='dotted', axis='y')
plt.show()
```

```python
#relation 2
#bar graph of star wars film franchise grouped by location

#here, by observing graph, we can conclude that, south atlantic have highest number of star trek franchise fan followed by 
#east north central and pacific
#while south central is the only region where number of star trek frenchise fan is greater than being not fan 

relation_chart2 = starWars.groupby(['Location (Census Region)',
             'Do you consider yourself to be a fan of the Star Trek franchise?']).size().unstack().plot(kind='bar', edgecolor='black', legend=True, stacked = False, color=('lightgreen', 'pink'), figsize=(10,10), fontsize=12)
plt.title('Fan of Star Trek Franchise by Location', fontsize=20)
relation_chart2.set_xlabel('Location', fontsize=20)
relation_chart2.set_ylabel("Frequency",fontsize=20)
plt.grid(color='grey', linewidth=3, linestyle='dotted', axis='y')
plt.show()
```

```python
#relation 3
#bar graph of expanded universe fan grouped by Age

#here, by observing graph, we can conclude that age group(18 - 29) has hoghest number of expanded universe fan
#age group( > 60) has lowest amount of expanded universe fan
#while age group(45 - 60) has highest number of respondent being not a fan of expanded universe
#age group(30 - 44) have equal number of repondent being fan of expanded unverse or being not

relation_chart3 = starWars.groupby(['Age',
             'Do you consider yourself to be a fan of the Expanded Universe?Œæ']).size().unstack().plot(kind='bar', edgecolor='black', legend=True, color=('lightgreen', 'orange'),figsize=(10,10), fontsize=12)
plt.title('Fan of Expanded Universe by Age', fontsize=20)
relation_chart3.set_xlabel('Age', fontsize=20)
relation_chart3.set_ylabel("Frequency",fontsize=20)
plt.grid(color='grey', linewidth=3, linestyle='dotted', axis='y')
plt.show()
```
